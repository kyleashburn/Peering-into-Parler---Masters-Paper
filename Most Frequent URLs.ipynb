```python
import json
import pandas as pd
```

```python
# first step is to check the length of the list
# each list can contain multiple dictionaries
# I need to cycle through the dictionaries within the list if it is longer than 1
# then, I need to pull each "u" from that dictionary
```

```python
# function to get the counts for domains in series passed into it
# pass in the series and it'll iterate through the whole thing
def count_urls(series) -> dict:
    url_dict = {}
    
    # kinda brute force but I'm not sure how to more efficiently do this
    for row in series:      
        # check if the item is a list
        if type(row) == list:
            
            # check the length of the list
            if len(row) > 1:
                               
                # cycling through each dict if more than one
                for entry in row:
                    
                    url = entry["u"]
                    
                    if url not in url_dict:
                        url_dict.setdefault(url, 1)
                
                    else:
                        url_dict[url] += 1
            
            # pull out the dict from inside it
            
            
            
            else:
                url = row[0]["u"]
            
                if url not in url_dict:
                    url_dict.setdefault(url, 1)
            
                else:
                    url_dict[url] += 1
                
    # returning the dictionary of terms 
    return url_dict
```

```python
# makes a function to pull the urls and sort them
def top_urls(url_df, url_dict):    
    # finding the top domains overall
    urls_count = count_urls(url_df["url"])
    
    url_dict.setdefault("overall", 
                       dict((sorted(urls_count.items(), key=lambda item: item[1],reverse=True))))    
                
    return url_dict
```

```python
# next step is to go ahead and enable me to loop through the 4 files
```

```python
file_list = ["min_act-urls.ndjson","mod_act-urls.ndjson","ver_act-urls.ndjson", "ext_act-urls.ndjson"]
```

```python
%%time 

url_dict = {}

for file in file_list:
    url_dict.setdefault(file, {})
    
    # reading in the file
    df = pd.read_json(file, lines=True)
    
    # filtering to just those having urls
    urls_df = df[~df["url"].isna()]
    
    print(file)
    # applying the function to pull the urls in each subcategory from the files & ordering them
    url_dict[file] = top_urls(urls_df, url_dict[file])
    
```

```python
%%time
# consolidating the url dict stuff from the subsection organization
url_consolidated = {}

for bucket in url_dict:
    for url in url_dict[bucket]["overall"]:
        if url not in url_consolidated:
            url_consolidated.setdefault(url, url_dict[bucket]["overall"][url])
            
        else:
            url_consolidated[url] += url_dict[bucket]["overall"][url]

```

```python
# sorting the dictionary
url_consolidated = dict(sorted(url_consolidated.items(), key=lambda item: item[1],reverse=True))
```

```python
# writing it to a json file
# opening the file we're working with here
url_counts = open("url_counts.json","a", encoding='utf-8')

#
url_counts.write(json.dumps(url_consolidated))

# closing the file
url_counts.close()
```
